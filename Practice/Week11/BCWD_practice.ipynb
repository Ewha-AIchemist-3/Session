## Effect size
- 요약 통계량의 하나입니다.
- 효과의 크기를 설명합니다. 두 그룹 간의 차이를 정량화하는 간단한 방법입니다.
- 다시 말해, 효과 크기는 차이의 크기를 강조합니다.
- 코헨의 효과 크기를 사용합니다.
- 코헨은 d(효과 크기)가 0.2이면 작은 효과 크기, d=0.5이면 중간 효과 크기, d=0.8이면 큰 효과 크기라고 제안했습니다.

## Relationship Between Variables
- 한 변수가 다른 변수에 대한 정보를 제공할 수 있다면, 두 변수는 서로 관련이 있다고 할 수 있습니다

## Hypothesis Testing
- Classical Hypothesis Testing
- 우리가 답하고자 하는 질문: "주어진 표본과 명백한 효과가 있을 때, 이러한 효과(cancer)가 우연히 발생할 확률은 얼마인가?"

1. 검정 통계량 선택
   - 명백한 효과의 크기를 수량화하는 첫 단계
   - 일반적으로 두 그룹 간의 평균 차이를 검정 통계량으로 선택

2. 귀무가설 정의
   - 명백한 효과가 실제로 존재하지 않는다는 가정에 기반한 모델
   - 귀무가설은 통계적으로 유의한 차이가 없다고 가정
   - 사람들이 반증하려고 하는 가설
   - 대립가설은 사람들이 증명하려고 하는 가설

3. p-값 계산
   - 귀무가설이 참일 때 관찰된 효과가 나타날 확률
   - p-값이 임계값보다 작거나 같으면 귀무가설을 기각
   - p-값이 낮으면 효과가 통계적으로 유의미하다고 판단
   - 즉, 우연히 발생했을 가능성이 낮고 더 큰 모집단에서도 나타날 가능성이 높음

예시:
   - 연구 질문: radius_mean과 area_mean이 서로 관련이 있는가?
   - 귀무가설: "종양 모집단에서 radius_mean과 area_mean의 관계는 0이다"
   - 이전 분석에서 이미 관계가 있다는 것을 알고 있지만, 통계적 검정을 통해 확인
   - p-값을 계산하여 귀무가설을 검정

```python
"""
실습 15: 가설 검정 수행하기

1. 귀무가설 설정
   H0: 종양 모집단에서 radius_mean과 area_mean의 관계는 0이다
   H1: 종양 모집단에서 radius_mean과 area_mean의 관계는 0이 아니다

2. 유의수준 설정
   α = 0.05 (일반적인 기준)

3. 검정 통계량 계산
"""

# scipy.stats를 사용한 대응표본 t-검정
statistic, p_value = stats.ttest_rel(# 여기에 첫 번째 변수,
                                   # 여기에 두 번째 변수)

# 결과 출력
print('검정 통계량:', statistic)
print('p-값:', p_value)

# 가설 검정 결과 해석
alpha = 0.05
if p_value < alpha:
    print('\n결과: 귀무가설 기각')
    print('radius_mean과 area_mean은 유의한 관계가 있습니다.')
else:
    print('\n결과: 귀무가설 채택')
    print('radius_mean과 area_mean의 관계가 통계적으로 유의하지 않습니다.')

"""
해석 가이드:
- p-값 < 0.05: 통계적으로 유의미
- p-값 < 0.01: 매우 유의미
- p-값 < 0.001: 극히 유의미
"""
```

```python
"""
실습 18: 교차 검증을 통한 모델 평가

1. 데이터 전처리
2. K-fold 교차 검증 수행
3. 성능 지표 계산 및 시각화
"""

# 필요한 라이브러리 임포트
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 준비
X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']

# 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 교차 검증 설정
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 교차 검증 수행
cv_scores = cross_val_score(rf_model, X_scaled, y, cv=kfold, scoring='accuracy')

# 결과 출력
print("\n교차 검증 결과:")
print(f"개별 폴드 정확도: {cv_scores}")
print(f"평균 정확도: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

# 다양한 하이퍼파라미터로 교차 검증
n_estimators_list = [50, 100, 200]
max_depth_list = [None, 5, 10]

results = []
for n_est in n_estimators_list:
    for max_d in max_depth_list:
        rf = RandomForestClassifier(n_estimators=n_est, 
                                  max_depth=max_d,
                                  random_state=42)
        scores = cross_val_score(rf, X_scaled, y, cv=kfold, scoring='accuracy')
        results.append({
            'n_estimators': n_est,
            'max_depth': max_d,
            'mean_score': scores.mean(),
            'std_score': scores.std()
        })

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)
print("\n하이퍼파라미터별 성능:")
print(results_df)
```
